#1. Análisis de facturación



```python
urls = {
    "Tienda 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "Tienda 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "Tienda 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "Tienda 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
}

```

```python

ingresos = {}
for tienda, url in urls.items():
  df = pd.read_csv(url)
  ingresos[tienda] = df["Precio"].sum()


for tienda, ingreso in ingresos.items():
  print(f"{tienda}: ${ingreso:,.2f}")
```

# 2. Ventas por categoría

```python


for tienda, url in urls.items():
 df = pd.read_csv(url)
 conteo = df['Categoría del Producto'].value_counts()

 print(f"\n{tienda} - Categorías más populares:")
 print(conteo)
```

# 3. Calificación promedio de la tienda


```python


for tienda, url in urls.items():
 df = pd.read_csv(url)
 promedio = df["Calificación"].mean()
 print(f"{tienda}: Calificación promedio = {promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt


for tienda, url in urls.items():
    df = pd.read_csv(url)
    conteo_productos = df['Producto'].value_counts()

    print(f"\n{tienda} - Productos más vendidos:")
    print(conteo_productos.head(5))

    print(f"\n{tienda} - Productos menos vendidos:")
    print(conteo_productos.tail(5))



for tienda, url in urls.items():
    df = pd.read_csv(url)
    conteo_productos = df['Producto'].value_counts()

    top_5 = conteo_productos.head(5)
    bottom_5 = conteo_productos.tail(5)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle(f"{tienda} - Productos más y menos vendidos", fontsize=16)

    top_5.plot(kind='bar', ax=axes[0], color='green')
    axes[0].set_title("Top 5 más vendidos")
    axes[0].set_ylabel("Cantidad vendida")
    axes[0].tick_params(axis='x', rotation=45)

    bottom_5.plot(kind='bar', ax=axes[1], color='red')
    axes[1].set_title("Top 5 menos vendidos")
    axes[1].set_ylabel("Cantidad vendida")
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

# 5. Envío promedio por tienda





```python


for tienda, url in urls.items():
 df = pd.read_csv(url)
 promedio = df["Costo de envío"].mean()
 print(f"{tienda}: Costo de envío promedio = ${promedio:,.2f}")
```

# 1. Gráfico de barras: Ingresos totales por tienda





```python
ingresos = {
    'Tienda 1': 1150880400,
    'Tienda 2': 1116343500,
    'Tienda 3': 1098019600,
    'Tienda 4': 1038375700
}

df_ingresos = pd.Series(ingresos)

plt.figure(figsize=(8,5))
df_ingresos.plot(kind='bar', color='skyblue')
plt.title('Ingreso Total por Tienda')
plt.ylabel('Ingreso ($)')
plt.xlabel('Tienda')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 2. Gráfico de líneas: Calificación promedio y costo de envío promedio por *tienda*

```python
calificaciones = {
    'Tienda 1': 3.98,
    'Tienda 2': 4.04,
    'Tienda 3': 4.05,
    'Tienda 4': 4.00
}

costos_envio = {
    'Tienda 1': 26018.61,
    'Tienda 2': 25216.24,
    'Tienda 3': 24805.68,
    'Tienda 4': 23459.46
}

df_calif = pd.Series(calificaciones)
df_envio = pd.Series(costos_envio)

plt.figure(figsize=(10,5))
plt.plot(df_calif.index, df_calif.values, marker='o', label='Calificación Promedio', color='blue')
plt.plot(df_envio.index, df_envio.values, marker='o', label='Costo de Envío Promedio', color='red')
plt.title('Calificación y Costo de Envío Promedio por Tienda')
plt.ylabel('Valor')
plt.xlabel('Tienda')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

*Ventas por Categoría:*

   La Tienda 3 es la que presenta mayor diversificación y mayor venta de productos claves.

La Tienda 1 obtiene la calificación más baja, lo que puede reflejar problemas en la atención al cliente o en la calidad de servicio.

*Productos Más y Menos Vendidos:*

La Tienda 1 depende en gran medida de pocos productos, lo que puede generar problemas de falta de stock o sobre stock de productos menos vendidos.

*Costo de Envío Promedio:*

La Tienda 1 presenta el costo de envío promedio más alto, mientras que la Tienda 4 muestra la mejor eficiencia en este aspecto.


**Conclusión**:

Pese a que la Tienda 1 presenta el mayor ingreso total, otros factores la hacen muy riesgoza, poco efciente en el corto y mediano plazo.

Por el contrario, la Tienda 3 muestra un desempeño más equilibrado, con mejores calificaciones, diversificación en categorías clave y costos de envío más razonables.
